# Futures Data

Futures contracts for an asset have different expiry dates. Usually, most trading activity is seen in the contract due to expire the earliest.

Successively expiring futures contracts are joined together to create continuous time-series data. This is done to make analysis and backtesting easy.

The notebook is divided into the following parts:

1. [Fetch Futures Data](#fetch)
1. [Plot the Data](#plot)

## Import Libraries

````python
# Import the library to fetch the data from Yahoo! Finance
import yfinance as yf

# Import matplotlib and set the style for plotting
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-darkgrid')
````

<a id='fetch'></a> 
## Fetch Futures Data

The continuous futures data is downloaded using the `yfinance` library.

Syntax: 
```python
import yfinance as yf
futures_data = yf.download(ticker_symbol, start=start_date, end=end_date)
```
Where
1. `ticker_symbol`: Ticker symbol whose data is required
2. `start_date`: The start date as a string in the format YYYY-MM-DD
3. `end_date`: The end date as a string in the format YYYY-MM-DD

Returns:

A DataFrame containing the historical data.

The `ticker_symbol` can be looked up on [Yahoo! Finance symbol lookup](https://finance.yahoo.com/lookup/).

For illustration, let us fetch the continuous lean hog futures with the symbol `HE=F`.

````python
# Fetch the futures data
futures_data = yf.download("HE=F", start="2017-01-01", end="2021-05-01")
````

<a id='plot'></a> 
## Plot the Data

The `Adj Close` price is plotted for better visualisation.

````python
# Plot the adjusted close price
(futures_data['Adj Close']).plot(figsize=(15, 7))

# Set title and labels for the plot
plt.title('Lean Hog Continuous Futures Adjusted Close Price', fontsize=14)
plt.xlabel('Year-Month', fontsize=12)
plt.ylabel('Price', fontsize=12)
plt.show()
````
